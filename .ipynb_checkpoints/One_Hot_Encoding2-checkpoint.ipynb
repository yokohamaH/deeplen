## kerasによるone-hotエンコーディングのサンプル
 - 英語バージョンだが同じことを日本語でもしたい
 

```python
from keras.preprocessing.text import Tokenizer

samples = ['The cat sat on the mat.', 'The dog ate my homework.']

tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(samples)

sequences = tokenizer.texts_to_sequences(samples)

one_hot_results = tokenizer.texts_to_matrix(samples, mode='binary')

word_index = tokenizer.word_index
print(word_index)
```

# 日本語でするone-hotエンコーディング
   - 今回は太宰治の「人間失格」を使用 

## まずはテキストデータの整理
 * 最初にいらない記号などを消す

```python
import sys
import re
 
path = 'debu/aozora/ningen_shikkaku.txt'
bindata = open(path, "rb")
lines = bindata.readlines()
for line in lines:
    text = line.decode('Shift_JIS')    # Shift_JISで読み込み
    text = re.split(r'\r',text)[0]     # 改行削除
    text = text.replace('｜','')       # ルビ前記号削除
    text = re.sub(r'《.+?》','',text)    # ルビ削除
    text = re.sub(r'［＃.+?］','',text)  # 入力者注削除
    print(text)
    file = open('new_ningen_shikkaku.txt','a',encoding='utf-8').write(text)  # UTF-8に変換
```

## いよいよエンコーディング
　　
  まず重複をなくし残ったものでone-hotエンコーディングを行う

```python
import pickle
f = open("./re_ningen_shikkaku.txt","rb")#list型としてload
lista = pickle.load(f)

chars = sorted(list(set(lista)))#set型は重複を許さないリスト

encode = []
dic = {}
dic2 = {}
dp = []
for i, v in enumerate(lista):
    
    if v in dp:
        encode.append(dp.index(v))
    else:
        encode.append(len(dp))
        dp.append(v)

for i , v in enumerate(chars):
    dic[v] = i
for i , v in enumerate(chars):
    dic2[i] = v

fs = open('re2_ningen_shikkaku.txt', 'wb')#辞書型の状態を保って保存
ft = open('dic_ningen_shikkaku.txt', 'wb')
ft2 = open('dic2_ningen_shikkaku.txt', 'wb')
fu = open('chars.txt', 'wb')

pickle.dump(encode, fs)
pickle.dump(dic, ft)
pickle.dump(dic2, ft2)
pickle.dump(chars, fu)
```
